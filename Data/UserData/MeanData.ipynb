```python
def process_and_normalize_tier_data(file_path, tier_name, is_scaled = False):
    """
    Process, normalize, and calculate mean metrics for the given TIER data using only 'win' == True rows.
    
    Parameters:
    - file_path: Path to the CSV file.
    - tier_name: Name of the TIER (e.g. 'CHALLENGER', 'GRANDMASTER').
    
    Returns:
    - DataFrame: Role mean metrics for the given TIER using only winning rows.
    """
    # Load the data
    df = pd.read_csv(file_path)
    
    # Adjust XPD@15 as described
    if(is_scaled == False):
        print('scaled == False')
        temp = []
        for i in range(df.shape[0]):
            if((i // 5) % 2 == 0):
                temp.append(df['XPD@15'][i] - df['XPD@15'][i+5])
            else:
                temp.append(df['XPD@15'][i] - df['XPD@15'][i-5])
        df['XPD@15'] = temp
    
    # Add 10000 to XPD@15 and normalize
    df['XPD@15'] += 10000
    
    # Filter rows where 'win' == True
    win_df = df[df['win'] == True]
    
    # Group by 'teamPosition' to get mean metrics
    role_mean_metrics = win_df.groupby('teamPosition')[['GPM', 'VSPM', 'DPM', 'KP ratio', 'XPD@15']].mean()
    
    # Rename the team positions and the column name for KP ratio
    role_mean_metrics = role_mean_metrics.rename(index={
        'MIDDLE': 'MID',
        'BOTTOM': 'ADC',
        'UTILITY': 'SUPPORT'
    })
    role_mean_metrics.rename(columns={'KP ratio': 'KP%'}, inplace=True)
    role_mean_metrics['TIER'] = tier_name
    
    return role_mean_metrics
```

```python
i_metrics = process_and_normalize_tier_data('https://raw.githubusercontent.com/wan6570/FlyAI_Project/Doyoon/API/LeagueofLegends/IRON.csv', 'IRON', is_scaled = True)
b_metrics = process_and_normalize_tier_data('https://raw.githubusercontent.com/wan6570/FlyAI_Project/Doyoon/API/LeagueofLegends/BRONZE.csv', 'BRONZE', is_scaled = True)
s_metrics = process_and_normalize_tier_data('https://raw.githubusercontent.com/wan6570/FlyAI_Project/Doyoon/API/LeagueofLegends/SILVER.csv', 'SILVER')
g_metrics = process_and_normalize_tier_data('https://raw.githubusercontent.com/wan6570/FlyAI_Project/Doyoon/API/LeagueofLegends/GOLD.csv', 'GOLD')
p_metrics = process_and_normalize_tier_data('https://raw.githubusercontent.com/wan6570/FlyAI_Project/Doyoon/API/LeagueofLegends/PLATINUM.csv', 'PLATINUM')
e_metrics = process_and_normalize_tier_data('https://raw.githubusercontent.com/wan6570/FlyAI_Project/Doyoon/API/LeagueofLegends/EMERALD.csv', 'EMERALD')
d_metrics = process_and_normalize_tier_data('https://raw.githubusercontent.com/wan6570/FlyAI_Project/Doyoon/API/LeagueofLegends/DIAMOND.csv', 'DIAMOND')
m_metrics = process_and_normalize_tier_data('https://raw.githubusercontent.com/wan6570/FlyAI_Project/Doyoon/API/LeagueofLegends/MASTER.csv', 'MASTER')
gm_metrics = process_and_normalize_tier_data('https://raw.githubusercontent.com/wan6570/FlyAI_Project/Doyoon/API/LeagueofLegends/GRANDMASTER.csv', 'GRANDMASTER')
challenger_metrics = process_and_normalize_tier_data('https://raw.githubusercontent.com/wan6570/FlyAI_Project/Doyoon/API/LeagueofLegends/CHALLENGER.csv', 'CHALLENGER')
```

```python
# Combine the metrics for GRANDMASTER and CHALLENGER
combined_metrics = pd.concat([i_metrics, b_metrics, s_metrics, g_metrics, p_metrics,e_metrics, d_metrics, m_metrics, gm_metrics, challenger_metrics], ignore_index=False)
combined_metrics
```
