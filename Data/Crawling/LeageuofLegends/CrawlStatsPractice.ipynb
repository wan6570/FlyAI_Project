```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests

# Load the CSV file into a DataFrame
df = pd.read_csv('final_links.csv', header=None)

# Convert the DataFrame back into a list of lists
final_links = df.values.tolist()

# Remove NaN values from the list of lists
final_links = [[link for link in links if pd.notna(link)] for links in final_links]

# Flatten the list of lists and reshape it to (-1, 1)
stats_links = np.array([link for links in final_links for link in links]).reshape(-1)

# Create an empty dataframe
final_df = pd.DataFrame()

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

first_link = stats_links[0]

requests.get(first_link, headers= headers).text
```

```python
# Make a request to the website
r = requests.get(first_link, headers= headers)
r_html = r.text

# Create a BeautifulSoup object and specify the parser
soup = BeautifulSoup(r_html, 'html.parser')

# Find the table
table = soup.find_all('table')[0]

# Convert the table to a string
table_str = str(table)

# Read the table into a dataframe
df = pd.read_html(table_str)[0]

# Transpose the dataframe and set the header row as the df header
df = df.T
df.columns = df.iloc[0]
df = df.iloc[1:]
```

```python
df.insert(1, 'Champion', 'Aatrox')
```

```python
# Assuming `soup` is your BeautifulSoup object
table_header = soup.find('table').find('thead').find_all('th')

additional_info = []

for th in table_header:
    img = th.find('img')  # Find 'img' within 'th'
    if img and img.has_attr('alt'):  # Check if 'img' exists and has 'alt' attribute
        additional_info.append(img['alt'])

print(additional_info)
```

```python
df['Champion']= additional_info
```
