```python
import pandas as pd
#pd.set_option('display.max_rows', None)
df =pd.read_csv('../../../ProcessedData/LCK_2023_SUMMER_WinLose.csv',index_col = 0)
df.isna().sum()
df.drop('CS in Enemy Jungle', axis = 1, inplace = True)
df.rename(columns = {"CS in Team's Jungle" : 'CS in Jungle'}, inplace=True)
df['Solo kills'].fillna(0, inplace=True)
```

```python
nan_df = (df.isna().sum() > 0)
nan_df_info = nan_df[nan_df == True]
```
