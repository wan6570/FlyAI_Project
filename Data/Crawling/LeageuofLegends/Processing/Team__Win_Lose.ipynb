```python
drop_less_than_10_df = df[df['Player'].isin(less_than_10.index[less_than_10 == False])]

drop_less_than_10_df
```

```python
top_df = drop_less_than_10_df[drop_less_than_10_df['Role'] == 'SUPPORT']
print(top_df['Player'].unique())
print(len(top_df['Player'].unique()))    
```

```python
team = {
    'KT' : ['Kiin', 'Cuzz', 'Bdd', 'Aiming', 'Lehends'],
    'GEN' : ['Doran', 'Peanut', 'Chovy', 'Peyz', 'Delight'],
    'HLE' : ['kingen', 'Grizzly', 'Clid', 'Zeka', 'Viper', 'Life'],
    'Dplus KIA' : ['Canna', 'Canyon', 'ShowMaker', 'Deft', 'Bible', 'Kellin'],
    'T1' : ['Zeus', 'Oner', 'Faker', 'Poby', 'Gumayusi', 'Keria'],
    'DRX' : ['Rascal', 'Croco','Juhan', 'FATE', 'Paduck', 'BeryL'],
    'LSB' : ['Burdol', 'Clear', 'Willer', 'Clozer', 'Teddy', 'Kael'],
    'BRION' : ['Morgan', 'UmTi', 'Karis', 'Hena', 'Effort'],
    'NS' : ['DnDn','Sylvie', 'FIESTA', 'Quad', 'Jiwoo', 'Peter'],
    'KDF' : ['DuDu', 'YoungJae', 'Bulldog', 'Taeyoon', 'Andil', 'Jun']
}
```

```python
player_to_team = {}
for team_name, players in team.items():
    for player in players:
        player_to_team[player] = team_name
    
print(player_to_team)
```

```python
drop_less_than_10_df.loc[:,'Team'] = drop_less_than_10_df['Player'].map(player_to_team)
```

```python
drop_less_than_10_df['Team'].value_counts()
```

```python
drop_less_than_10_df.to_csv('../../../ProcessedData/LCK_2023_SUMMER_AddTeam_DropSubs.csv', index=False)
```
