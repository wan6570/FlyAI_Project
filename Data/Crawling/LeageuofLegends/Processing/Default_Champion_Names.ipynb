```python
import pandas as pd
import numpy as np

df = pd.read_csv('../../../Data/RawData/GameStats/LCK_2023_SUMMER.csv')
df
```

```python
df['Champion'].unique()
```

```python
target_dict = {
    'Kai' : "Kai'Sa",
    'K' : "K'Sante",
    'Kog' : "Kog'Maw",
    'Kha' : "Kha'Zix",
    'Rek' : "Rek'Sai",
    'Cho' : "Cho'Gath", 
}

def replace_champion_name(champion_name):
    if champion_name in target_dict.keys():
        return target_dict[champion_name]
    else:
        return champion_name

def replace_champion_name_list(champion_name_list):
    return [replace_champion_name(champion_name) for champion_name in champion_name_list]

def replace_champion_name_df(df):
    df['Champion'] = replace_champion_name_list(df['Champion'])
    return df

replace_champion_name_df = replace_champion_name_df(df)
replace_champion_name_df['Champion'].unique()

replace_champion_name_df.to_csv('../../../ProcessedData/Complete_Chmapion_Name.csv', index=False)
```
