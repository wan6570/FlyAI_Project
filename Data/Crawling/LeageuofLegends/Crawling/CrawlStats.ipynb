```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests

df = pd.read_csv('final_links_2023_Summer.csv', header=None)
final_links = df.values.tolist()
final_links = [[link for link in links if pd.notna(link)] for links in final_links]

stats_links = np.array([link for links in final_links for link in links]).reshape(-1)

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

print(stats_links)
```

```python
final_df = pd.DataFrame()

for link in stats_links:
    r = requests.get(link, headers= headers)
    r_html = r.text

    soup = BeautifulSoup(r_html, 'html.parser')

    table = soup.find_all('table')[0]
    table_str = str(table)
    
    df = pd.read_html(table_str)[0]

    df = df.T
    df.columns = df.iloc[0]
    df = df.iloc[1:]

    # df에 챔피언 열 추가
    df.insert(1, 'Champion', 'Aatrox')

    table_header = soup.find('table').find('thead').find_all('th')

    additional_info = []

    for th in table_header:
        img = th.find('img')  
        if img and img.has_attr('alt'):
            additional_info.append(img['alt'])
    
    df['Champion']= additional_info

    final_df = pd.concat([final_df, df])

final_df.reset_index(drop=True, inplace=True)
print(final_df)
```

```python
LCK_2023_SPRING_DataFrame = pd.DataFrame(final_df)
```

```python
LCK_2023_SPRING_DataFrame.to_csv('LCK_2023_SUMMER_DataFrame.csv', index=False)
```
