```python
import requests
from bs4 import BeautifulSoup
import time

url = "https://gol.gg/tournament/tournament-matchlist/LCK%20Summer%202023/"
base_url = "https://"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table')

links = []
for tr in table.tbody.find_all('tr'):
    a = tr.find('a')
    if a:
        link = a.get('href')
        links.append(base_url + 'gol.gg' + link[2:]) 

print(links)
```

```python
valid_links = []
for link in links:
    temp = []
    base_game_url = link.replace('page-summary', 'page-fullstats')  
    game_number = int(base_game_url.split('/')[-3])  
    for i in range(3):  
        game_url = base_game_url.replace(str(game_number), str(game_number + i))
        response = requests.get(game_url, headers=headers)
        if response.status_code == 200: 
            temp.append(game_url)
        time.sleep(3)
    valid_links.append(temp)
```

```python
import pandas as pd

df = pd.DataFrame(valid_links)
df.to_csv('final_links.csv', index=False, header=False)
```
