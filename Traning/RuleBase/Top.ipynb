```python
import pandas as pd

df =pd.read_csv('../../Data/ProcessedData/LCK_2023_SUMMER_DROP_NAN.csv')

player_counts = df['Player'].value_counts()
players_to_keep = player_counts[player_counts > 25].index
df = df[df['Player'].isin(players_to_keep)]

Top_df = df[df['Role'] == 'TOP']
Win_Top_df = Top_df[Top_df['Result'] == 'Win']
Lose_Top_df = Top_df[Top_df['Result'] == 'Lose']
```

```python
Top_df['Player'].value_counts()
```

```python
non_object_columns = Win_Top_df.select_dtypes(exclude=['object']).columns
Win_Top_df_avg = Win_Top_df.groupby('Player')[non_object_columns].mean()

Win_Top_df_avg['DMG%']
```

```python
avg_win_stats = (Win_Top_df_avg / Win_Top_df[non_object_columns].mean())
avg_stats = Top_df.groupby('Player')[non_object_columns].mean() / Top_df[non_object_columns].mean()
avg_lose_stats = Lose_Top_df.groupby('Player')[non_object_columns].mean() / Lose_Top_df[non_object_columns].mean()
```
